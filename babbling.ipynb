```python
import math
import pandas as pd
%matplotlib inline
```

```python
k_contigency = 500/math.log(3000)
print(k_contigency)
k_nonContigency = 600/math.log(3000)
print(k_nonContigency)
k_control = 70
```

```python
# Experiment 1 
# k_contigency, k_nonContigency and k_control is same
k_contigency = k_nonContigency = k_control = 60
```

```python
numContigency = k_contigency * math.log(6000)
print(numContigency)
```

```python
numNonContigency = k_nonContigency * math.log(20000)
print(numNonContigency)
```

```python
numControl = k_control * math.log(26000)
print(numControl)
```

```python
def expectedUniqueN(n, k):
    return n * ( 1- pow((n-1)*1.0/n,k))
```

```python
EX_contigency = [expectedUniqueN(numContigency, i) for i in range(100,3000,100)]
EX_noncontigency = [expectedUniqueN(numNonContigency, i) for i in range(100,3000,100)]
df = pd.DataFrame(data={'contigency': EX_contigency,
                        'noncontigency': EX_noncontigency}, 
                  index=range(100,3000,100))
df.plot()
```

```python
# Now let's build a control dataset 
# The control has a 762 words outof 26k corpus

#if we sample 6000 words from this list it will contain 
numContigencyControl = expectedUniqueN(numControl, 6000)
print(numContigencyControl)

#if we sample 6000 words from this list it will contain 
numNonContigencyControl = expectedUniqueN(numControl, 20000)
print(numNonContigencyControl)
```

```python
EX_contigency_control = [expectedUniqueN(numContigencyControl, i) for i in range(100,3000,100)]
EX_noncontigency_control = [expectedUniqueN(numNonContigencyControl, i) for i in range(100,3000,100)]
df_control = pd.DataFrame(data={'contigency control': EX_contigency_control,
                        'noncontigency control': EX_noncontigency_control}, 
                  index=range(100,3000,100))
df_control.plot()
```

```python
EX_contigency_diff = [expectedUniqueN(numContigency, i) - expectedUniqueN(numControl, i) for i in range(100,3000,100)]
EX_noncontigency_diff = [expectedUniqueN(numNonContigency, i) - expectedUniqueN(numControl, i) for i in range(100,3000,100)]
df_control_diff = pd.DataFrame(data={'contigency diff': EX_contigency_diff,
                        'noncontigency diff': EX_noncontigency_diff}, 
                  index=range(100,3000,100))
df_control_diff.plot()
```
